1. What Python Is & Why It Is Popular?  
:- Python is a high-level, general-purpose programming language designed for readability.It is popular because:Simple Syntax: Code looks like plain English.Versatile: Used in AI, web development, and data science.Large Community: Vast libraries prevent writing code from scratch

2. What is an interpreter in Python ?
:- An interpreter is a program that executes code line by line. It translates Python code directly into machine-readable instructions without requiring a separate compilation step.

3. What are pre-defined keywords in Python ?
:- Keywords are reserved words that define the syntax and structure of Python. Examples include if, for, def, return, and True

4. Can keywords be used as variable names ?
:- No. Keywords cannot be used as variable names. Doing so will trigger a SyntaxError.

5. What is mutability in Python ?
:- Mutability refers to whether an object's value can be changed after it is created.Mutable: Objects can be modified in place.Immutable: Objects cannot be altered after creation.

6. Why are lists mutable, but tuples are immutable ?
:- Lists (mutable): Designed for dynamic collections where elements change, grow, or shrink.
   Tuples (immutable): Designed for fixed data sets, ensuring data safety and faster performance

7. What is the difference between “==” and “is” operators in Python ?
:- == Operator: Checks if the values of two objects are equal.
  is Operator: Checks if two variables point to the exact same memory location.

8. What are logical operators in Python ?
:- Logical operators combine conditional statements:
 and: Returns True if both statements are true.
 or: Returns True if at least one statement is true.
 not: Reverses the logical state (turns True to False).

9. What is type casting in Python ?
:-  Type casting is the process of converting a variable from one data type to another, such as turning a string into an integer.

10. What is the difference between implicit and explicit type casting ?
:- Implicit: Python automatically converts  types (e.g., adding an integer to a float turns the result into a float).
   Explicit: The programmer manually forces a conversion using functions like int(), str(), or list().

11.  What is the purpose of conditional statements in Python ?
:- Conditional statements (if, elif, else) control the flow of a program. They allow the code to execute specific blocks only when certain conditions are met.

12. How does the elif statement work ?
:- The elif (short for else-if) statement checks a new condition only if the previous if or elif conditions evaluated to False. It prevents unnecessary checks once a match is found.

13. What is the difference between for and while loops ?
:- For Loop: Iterates over a sequence a predetermined number of times.
   While Loop: Repeats continuously as long as a specific condition remains true.

14. Describe a scenario where a while loop is more suitable than a for loop?
:- A while loop is ideal when you do not know how many iterations are needed in advance.Example: Prompting a user to enter their password until they type the correct one.


````python
1.Write a Python program to print "Hello, World!y "
:- print("Hello, World!")
2.Write a Python program that displays your name and age8.
:- name = input("Enter your name: ")
age = input("Enter your age: ")
print(f"Hello, {name}! You are {age} years old.")

3.Write code to print all the pre-defined keywords in Python using the keyword library.
:- import keyword

# Get the list of all Python keywords
keywords_list = keyword.kwlist

# Print each keyword
for word in keywords_list:
    print(word)

# Print the total count of keywords
print("\nTotal number of keywords:", len(keywords_list))

4. Write a program that checks if a given word is a Python keyword.
:-  import keyword


def check_python_keyword(word):
    # Check if the word is a Python keyword
    if keyword.iskeyword(word):
        print(f"'{word}' is a Python keyword.")
    else:
        print(f"'{word}' is NOT a Python keyword.")


# Test the program
test_word = input("Enter a word to check: ")
check_python_keyword(test_word)

5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently
for each.
:- my_list = ["apple", "banana", "cherry"]
my_tuple = ("apple", "banana", "cherry")

print("Original List:", my_list)
print("Original Tuple:", my_tuple)
print("-" * 40)

6. Write a function to demonstrate the behavior of mutable and immutable arguments.
:-
 def demonstrate_arguments(immutable_num, mutable_list):
    print("--- Inside Function (Start) ---")

    # 1. Modify the immutable argument
    immutable_num += 10
    print(f"Modified local number: {immutable_num}")

    # 2. Modify the mutable argument
    mutable_list.append("new_item")
    print(f"Modified local list: {mutable_list}")
    print("--- Inside Function (End) ---\n")

7.  Write a program that performs basic arithmetic operations on two user-input numbers.
:-
  # 1. Take input from the user and convert to floats
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

# 2. Perform arithmetic operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2

# 3. Handle division carefully to avoid crashes
if num2 != 0:
    division = num1 / num2
else:
    division = "Undefined (Cannot divide by zero)"

# 4. Display the results
print("\n--- Results ---")
print(f"{num1} + {num2} = {addition}")
print(f"{num1} - {num2} = {subtraction}")
print(f"{num1} * {num2} = {multiplication}")
print(f"{num1} / {num2} = {division}")

8. Write a program to demonstrate the use of logical operators.
:-
   # Setup test scenarios
has_high_gpa = True
has_good_conduct = False
has_financial_need = True

print("=== Variable Status ===")
print(f"High GPA: {has_high_gpa}")
print(f"Good Conduct: {has_good_conduct}")
print(f"Financial Need: {has_financial_need}\n")

# 1. Demonstrating 'and' (Both conditions MUST be True)
# Check for academic excellence award
academic_award = has_high_gpa and has_good_conduct
print(f"1. [AND] High GPA AND Good Conduct? -> {academic_award}")

# 2. Demonstrating 'or' (At least ONE condition MUST be True)
# Check for general scholarship eligibility
scholarship_eligible = has_high_gpa or has_financial_need
print(
    f"2. [OR] High GPA OR Financial Need? -> {scholarship_eligible}"
)

# 3. Demonstrating 'not' (Reverses the Boolean value)
# Check if student is banned (Opposite of good conduct)
is_banned = not has_good_conduct
print(f"3. [NOT] Banned (Not Good Conduct)? -> {is_banned}")

9. Write a Python program to convert user input from string to integer, float, and boolean types.
:-
  # 1. Get string input from the user
user_input = input("Enter a value to convert (e.g., 1, 3.14, True): ")
print(f"\nOriginal Input: '{user_input}' (Type: {type(user_input).__name__})")
print("-" * 50)

# 2. Convert to Integer
try:
    int_value = int(user_input)
    print(f"As Integer: {int_value} (Type: {type(int_value).__name__})")
except ValueError:
    print("As Integer: Conversion failed (Not a valid whole number)")

# 3. Convert to Float
try:
    float_value = float(user_input)
    print(f"As Float:   {float_value} (Type: {type(float_value).__name__})")
except ValueError:
    print("As Float:   Conversion failed (Not a valid decimal number)")

# 4. Convert to Boolean safely
# Note: Simply using bool(user_input) returns True for ANY non-empty string.
# We check the actual word value for an accurate logical conversion.
truthy_values = ["true", "1", "yes", "on", "t"]
falsy_values = ["false", "0", "no", "off", "f"]

cleaned_input = user_input.strip().lower()

if cleaned_input in truthy_values:
    bool_value = True
    print(f"As Boolean: {bool_value} (Type: {type(bool_value).__name__})")
elif cleaned_input in falsy_values:
    bool_value = False
    print(f"As Boolean: {bool_value} (Type: {type(bool_value).__name__})")
else:
    print("As Boolean: Conversion ambiguous (Not recognized as True/False)")

10. Write code to demonstrate type casting with list elements.
:-
  # 1. Create a list of mixed data types or string data
string_prices = ["10.50", "20", "5.75", "100"]
print(f"Original List: {string_prices}")
print(f"Original Element Types: {[type(x).__name__ for x in string_prices]}\n")

# 2. Type casting elements to FLOATS using a list comprehension
float_prices = [float(item) for item in string_prices]
print(f"Casted to Floats: {float_prices}")
print(f"New Element Types: {[type(x).__name__ for x in float_prices]}\n")

# 3. Type casting float elements to INTEGERS (truncating decimals)
int_prices = [int(item) for item in float_prices]
print(f"Casted to Integers: {int_prices}")
print(f"New Element Types: {[type(x).__name__ for x in int_prices]}\n")

# 4. Alternative: Type casting using the map() function
# This converts integers back into strings
string_again = list(map(str, int_prices))
print(f"Casted back to Strings: {string_again}")
print(f"Final Element Types: {[type(x).__name__ for x in string_again]}")

11. Write a program that checks if a number is positive, negative, or zero.
:-
  # 1. Take a number as input from the user
number = float(input("Enter a number: "))

# 2. Check the condition using if-elif-else
if number > 0:
    print(f"{number} is a positive number.")
elif number < 0:
    print(f"{number} is a negative number.")
else:
    print("The number is zero.")

12. Write a for loop to print numbers from 1 to 10
:-
   # Loop from 1 up to (but not including) 11
for i in range(1, 11):
    print(i)

13. Write a Python program to find the sum of all even numbers between 1 and 50
:-
   even_sum = sum(range(2, 51, 2))
print("The sum of all even numbers between 1 and 50 is:", even_sum)

14.Write a program to reverse a string using a while loop.
:-
   # 1. Take a string input from the user
original_string = input("Enter a string to reverse: ")

# 2. Initialize variables
reversed_string = ""
# Start at the last index of the string
index = len(original_string) - 1

# 3. Loop backwards through the string
while index >= 0:
    # Append the character to the new string
    reversed_string += original_string[index]
    # Move to the previous character
    index -= 1

# 4. Display the result
print("Reversed string:", reversed_string)
15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
:-
  # 1. Take a whole number input from the user
num = int(input("Enter a positive integer: "))

# 2. Store the original number for the final print statement
original_num = num

# 3. Initialize the factorial result variable to 1
factorial = 1

# 4. Handle negative numbers, zero, and positive numbers
if num < 0:
    print("Factorial does not exist for negative numbers.")
elif num == 0:
    print("The factorial of 0 is 1.")
else:
    # Loop to multiply the numbers downward
    while num > 1:
        factorial *= num  # Multiply current number to the result
        num -= 1  # Decrease the number by 1

    # 5. Display the final result
    print(f"The factorial of {original_num} is {factorial}")
````
